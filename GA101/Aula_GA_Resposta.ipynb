### 4.0 Construção da População

```python
#---FUNÇÃO QUE GERA UMA POPUÇÃO

def gera_pop(tamanho_populacao, Objetivo, dominio):
    
    #Lista que guardará os indivíduos da população
    pop = []

    
    #Construindo a população
    for cont in range(tamanho_populacao):
        
        #Definindo o indivíduo
        individuo = random.choices(dominio, k= len(Objetivo))
        
        #Inserindo o individuo na pop
        pop.append(individuo)
        
        
    return pop
```

### 5.0 Calculo do Fitness

```python
#---FUNÇAÕ QUE CONPUTA A O FITNESS DOS INDIVIUDUOS DE UMA POPULAÇÃO

def fitness(pop, Objetivo):
    
    
    #Lista contendo o fitness dos integrantes de uma população
    fit = []
    
    #Calculo do fitness
    for cont_1 in pop:
        
        pontuacao = 0         
        
        for cont_2, cont_3 in zip(cont_1,Objetivo):
            
            #Conparando o individuo com o objetivo
            
            if cont_2 == cont_3:
                
                pontuacao += 1
            
            
        fit.append(pontuacao)
        
    
    #Retornar a lista de fitness
    return fit
```

### 6.0 Mostrar melhor resposta

```python
#---CONSTRUIR UMA FUNÇÃO QUE MOSTRA A MELHOR RESPOSTA

def apresentacao(max_ind):
    
    melhor =''
    for cont in max_ind:
        melhor += cont
        
    return melhor    
```

## Encontrar a frase através de um algoritmo genético

### 9.0 Seleção dos indivíduos

```python
#---FUNÇÃO DE SELEÇÃO DOS INDIVÍDUOS por torneio

def selecao(pop,fit, N, q):
    
    nova_pop = []
    
    #--Guardando o melhor indice
    
    #Obter o melhor fitness
    melhor_fit_indice = fit.index(max(fit))
    
    #Inserindo o melhor fitness na população
    nova_pop.append(pop[melhor_fit_indice])
    
    
    #Realizando os torneios
    valores_possiveis = list(range(len(pop)))   
    
    
    for cont in range(N-1):
        
        #Escolhendo os competidores da rodada
        competidores = random.choices(valores_possiveis, k=q)
        
        #Max valore dos competidores
        Max = 0.0
        
        indice = 'Null'
        
        for cont_2 in competidores:
            
            if fit[cont_2] >= Max:
                
                Max    = fit[cont_2]
                
                indice = cont_2
                
                
        #Atualizando a nova população
        nova_pop.append(pop[indice])
        
        
    return nova_pop
```

### 10.0 Função para realizar o Crossing-Over

```python
def cross_over(pop,Objetivo):
    
    
    for cont_1, cont_2 in zip(range(0,len(pop),2), range(1,len(pop),2)):
        
        
        filho_1 = []
        
        filho_2 = []
        
        
        for cont_3 in range(len(Objetivo)):
            
            if cont_3 < (len(Objetivo)/2.0):
                
                filho_1.append(pop[cont_1][cont_3])
                
                filho_2.append(pop[cont_2][cont_3])
                
            else:
                
                filho_1.append(pop[cont_2][cont_3])
                
                filho_2.append(pop[cont_1][cont_3])
                
                
        pop.append(filho_1)
        pop.append(filho_2)
                
                
                
    return pop
```

### 11.0 Função para realizar Mutação

```python
def mutacao(pop,fit, dominio, prob_mutacao):
    
    nova_pop = []
    
    
    #--Quandano o melhor fitness
    
    #Obter o melhor fitness
    melhor_fit_indice = fit.index(max(fit))
    
    #Inserindo o melhor fitness na população
    nova_pop.append(pop[melhor_fit_indice])
    
    
    for cont_1  in range(1,len(pop)):
        
        temp = []
        
        for cont_2 in range(len(pop[cont_1])):

            
            if random.random() <= prob_mutacao:
                
                temp.append(random.choices(dominio, k= 1)[0])
                
            else:
                
                temp.append(pop[cont_1][cont_2])
            
        
        
        #Atualizando a nova população
        nova_pop.append(temp)
        
                
    return nova_pop
    
```

## 12.0 Busca pelo algoritmo genético

#####   Características da população e do procedimento de busca aleatória

```python
#---DEFININDO CARACTERISTICAS DA POPULAÇÃO 
tamanho_populacao = 100

geracoes = 200

#---Definindo as propriendades da seleção

N = tamanho_populacao  #Número de torneios realizados

q = 5              #Númeo de competidores em um torneio


#---Definindo parametros da mutação
prob_mutacao = 0.05  #Probabilidade de se ter uma mutacao
```

# -----------------------------------------------------------------------------------------

## 13.0 Realizando a busca por um algoritmo genético

```python
#---Inicializando a população
pop = gera_pop(tamanho_populacao, Objetivo, dominio)

#--Contabilizando o fitness da população
fit = fitness(pop, Objetivo)

#--Obter o melhor fitness
melhor_fit_indice = fit.index(max(fit))

#--Mostrar o melhor resultados
melhor_ind = apresentacao(pop[melhor_fit_indice])


#--Mostrando o melhor resultados
print('Geracao. inicial --> {} --> Fitness {}'.format(melhor_ind,fit[melhor_fit_indice]))
    
print()



for cont in range(1,geracoes+1):
    
    #--Seleção dos melhores indivíduos (Obtendo nova geração)
    pop = selecao(pop,fit, N, q)
    

    #--Cross-over
    pop = cross_over(pop,Objetivo)
    
    #--Inserindo mutação na população
    pop = mutacao(pop,fit, dominio, prob_mutacao)
    

    #--Contabilizando o fitness da população
    fit = fitness(pop, Objetivo)

    
    #---------------------------------------------------------------------
    #--------------------Apresentados os resultados-----------------------
    #---------------------------------------------------------------------
        
    
    #--Obter o melhor fitness
    melhor_fit_indice = fit.index(max(fit))    
    
    
    #--Mostrar o melhor resultados
    melhor_ind = apresentacao(pop[melhor_fit_indice])
    
    
    #--Mostrando o melhor resultados
    print('Gera. {} --> {} --> Fitness {}'.format(cont, melhor_ind,fit[melhor_fit_indice]))
    
    print()
    
    time.sleep(0.4)
    


```
